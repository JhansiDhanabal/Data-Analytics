number of loans issued changed over the years

seasonal variation in loan issuance

```python
df['issue_date'] = pd.to_datetime(df['issue_date'])

# Extract year from the 'issue_d' column
df['issue_year'] = df['issue_date'].dt.year

# Count the number of loans issued each year
loans_per_year = df['issue_year'].value_counts().sort_index()

# Plot the number of loans issued over the years
plt.figure(figsize=(10, 6))
loans_per_year.plot(kind='bar', color='skyblue')
plt.title('Number of Loans Issued Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Loans Issued')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#Seasonal Variation in Loan Issurance
# Extract month from the 'issue_d' column
df['issue_month'] = df['issue_date'].dt.month

# Count the number of loans issued each month
loans_per_month = df.groupby('issue_month').size()

# Plot the seasonal variation in loan issuance
plt.figure(figsize=(10, 6))
loans_per_month.plot(kind='line', marker='o', color='green')
plt.title('Seasonal Variation in Loan Issuance')
plt.xlabel('Month')
plt.ylabel('Number of Loans Issued')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Correlation between loan amount interset rate

```python
correlation = df['loan_amount'].corr(df['interest_rate'])

print("Correlation between loan amount and interest rate:", correlation)
```

```python
# Calculate the correlation between borrower's income and loan amount
income_loan_correlation = df['annual_inc'].corr(df['loan_amount'])

# Calculate the correlation between borrower's income and interest rate
income_interest_correlation = df['annual_inc'].corr(df['interest_rate'])

print("Correlation between borrower's income and loan amount:", income_loan_correlation)
print("Correlation between borrower's income and interest rate:", income_interest_correlation)
```

```python
# Extract year from the 'issue_date' column
df['issue_year'] = df['issue_date'].dt.year

# Plot the trend between loan availability and income earned by the bank over time
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='issue_year', y='loan_amount', marker='o', label='Loan Availability')
sns.lineplot(data=df, x='issue_year', y='annual_inc', marker='o', label='Bank Income')
plt.title('Trend between Loan Amount and Bank Income')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Relationship Between Interest Rate and Loan Grade

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='grade', y='interest_rate', data=df, order=sorted(df['grade'].unique()))
plt.title('Relationship Between Interest Rate and Loan Grade')
plt.xlabel('Loan Grade')
plt.ylabel('Interest Rate')
plt.grid(True)
plt.tight_layout()
plt.show()
```
